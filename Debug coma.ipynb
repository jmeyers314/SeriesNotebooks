```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import galsim
```

```python
a=galsim.LinearOpticalet(1.0,3,1,0,0)
b=galsim.Gaussian(sigma=1.e-8)
c=galsim.Convolve(a,b)
img1=a.drawImage(nx=100, ny=100, scale=0.1, method='no_pixel')
img2=c.drawImage(nx=100, ny=100, scale=0.1, method='fft')
plt.imshow(img1.array)
plt.colorbar()
plt.show()
plt.imshow(img2.array)
plt.colorbar()
```
