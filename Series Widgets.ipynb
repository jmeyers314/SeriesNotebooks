```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
import numpy as np
import galsim

from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML
```

```python
def img_and_colorbar(fig, grid):
    inner_grid = gs.GridSpecFromSubplotSpec(100, 100, subplot_spec=grid, wspace=0, hspace=0)
    img_ax = fig.add_subplot(inner_grid[:,0:90])
    cbar_ax = fig.add_subplot(inner_grid[:,91:95])
    return img_ax, cbar_ax

def plot(array1, array2, vmin=None, vmax=None):
    fig = plt.figure(figsize=(15,3))
    grid = gs.GridSpec(1,17)

    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,0:5])
    im = img_ax.imshow(array1, vmin=vmin, vmax=vmax)
    plt.colorbar(im, cax=cbar_ax)
    img_ax.set_title("Series profile")

    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,6:11])
    im = img_ax.imshow(array2, vmin=vmin, vmax=vmax)
    plt.colorbar(im, cax=cbar_ax)
    img_ax.set_title("Exact profile")

    tmp1 = array1/array1.sum()
    tmp2 = array2/array2.sum()
    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,12:17])
    im = img_ax.imshow((tmp1-tmp2)/tmp2.max())
    plt.colorbar(im, cax=cbar_ax)    
    img_ax.set_title("normalized residual")
```

```python
def draw_spergel(nu=0.0, mu=1.0, epsilon=0.0, phi0=0.0, jmax=4):
    gal_exact = galsim.Spergel(nu=nu, scale_radius=1.0).dilate(mu).shear(e=epsilon, beta=phi0*galsim.radians)
    gal_series = galsim.SpergelSeries(nu=nu, scale_radius=1.0, jmax=jmax).dilate(mu).shear(e=epsilon, beta=phi0*galsim.radians)
    im_exact = gal_exact.drawImage(nx=32, ny=32, scale=0.2)
    im_series = gal_series.drawImage(nx=32, ny=32, scale=0.2)
    plot(im_series.array, im_exact.array)
```

```python
interact(draw_spergel,
         nu=(-0.5, 4.0, 0.01),
         mu=(0.8, 1.2, 0.01),
         epsilon=(0.0, 0.99,0.01),
         phi0=(0.0, np.pi, 0.01),
         jmax=(4,12)
         )
```

```python
def draw_moffat(beta=3.0, mu=1.0, epsilon=0.0, phi0=0.0, jmax=4):
    psf_exact = galsim.Moffat(beta=beta, scale_radius=1.0).dilate(mu).shear(e=epsilon, beta=phi0*galsim.radians)
    psf_series = galsim.MoffatSeries(beta=beta, scale_radius=1.0, jmax=jmax).dilate(mu).shear(e=epsilon, beta=phi0*galsim.radians)
    im_exact = psf_exact.drawImage(nx=32, ny=32, scale=0.2)
    im_series = psf_series.drawImage(nx=32, ny=32, scale=0.2)
    plot(im_series.array, im_exact.array)
```

```python
interact(draw_moffat,
         beta=(1.1, 5.0, 0.01),
         mu=(0.8, 1.2, 0.01),
         epsilon=(0.0, 0.1,0.001),
         phi0=(0.0, np.pi, 0.01),
         jmax=(4,12)
         )
```

```python
def draw_convolution(nu=0.5, beta=4.5, psf_mu=1.0, psf_epsilon=0.0, psf_phi0=0.0, 
                     gal_mu=1.0, gal_epsilon=0.0, gal_phi0=0.0, psf_jmax=4, gal_jmax=8):
    psf_exact = galsim.Moffat(beta=beta, scale_radius=1.0).dilate(psf_mu).shear(e=psf_epsilon, beta=psf_phi0*galsim.radians)
    psf_series = galsim.MoffatSeries(beta=beta, scale_radius=1.0, jmax=psf_jmax).dilate(psf_mu).shear(e=psf_epsilon, beta=psf_phi0*galsim.radians)
    gal_exact = galsim.Spergel(nu=nu, scale_radius=1.0).dilate(gal_mu).shear(e=gal_epsilon, beta=gal_phi0*galsim.radians)
    gal_series = galsim.SpergelSeries(nu=nu, scale_radius=1.0, jmax=gal_jmax).dilate(gal_mu).shear(e=gal_epsilon, beta=gal_phi0*galsim.radians)
    conv_exact = galsim.Convolve(psf_exact, gal_exact)
    conv_series = galsim.SeriesConvolution(psf_series, gal_series)
    im_exact = conv_exact.drawImage(nx=32, ny=32, scale=0.2)
    im_series = conv_series.drawImage(nx=32, ny=32, scale=0.2)
    plot(im_series.array, im_exact.array)
```

```python
interact(draw_convolution,
         psf_mu=(0.8, 1.2, 0.01),
         psf_epsilon=(0.0, 0.1, 0.001),
         psf_phi0=(0.0, np.pi, 0.01),
         gal_mu=(0.8, 1.2, 0.01),
         gal_epsilon=(0.0, 0.8, 0.01),
         gal_phi0=(0.0, np.pi, 0.01),
         psf_jmax=(4,8),
         gal_jmax=(8,12)
         )
```
