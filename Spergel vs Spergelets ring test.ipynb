```python
import numpy as np
import math
import lmfit
import galsim
```

```python
def shearme(e1, e2, g1, g2):
    e = e1 + 1j*e2
    g = g1 + 1j*g2
    eout = (e + g)/(1+g.conjugate()*e)
    return eout.real, eout.imag
```

```python
def drawtrue(e1, e2, HLR, nu, psffwhm):
    psf = galsim.Gaussian(fwhm=psffwhm)
    gal = galsim.Spergel(nu=nu, half_light_radius=HLR).shear(e1=e1, e2=e2)
    conv = galsim.Convolve(psf, gal)
    return conv.drawImage(nx=32, ny=32, scale=0.2)
```

```python
def resid(params, nu, jmax, psffwhm, im):
    flux = params['flux'].value
    HLR = params['HLR'].value
    e1 = params['e1'].value
    e2 = params['e2'].value
    
    psf = galsim.Gaussian(fwhm=psffwhm)
    gal = galsim.SpergelSeries(nu=nu, half_light_radius=HLR, jmax=jmax).shear(e1=e1, e2=e2)
    conv = galsim.SeriesConvolution(psf, gal)
    model = conv.drawImage(nx=32, ny=32, scale=0.2)
    return (model.array - im.array).ravel()

def fitme(im, e1, e2, HLR, nu, jmax, psffwhm):
    params = lmfit.Parameters()
    params.add('flux', 1.0, min=0.0)
    params.add('HLR', HLR)
    params.add('e1', e1)
    params.add('e2', e2)

    result = lmfit.minimize(resid, params, args=(nu, jmax, psffwhm, im))
    return result.params['e1'].value, result.params['e2'].value
```

```python
e = 0.5
HLR = 0.4
psffwhm=0.6
nu=-0.5
jmax=8

g1s = [0.0, 0.01]
g2s = [0.0, 0.02]
g1hats = []
g2hats = []
for g1, g2 in zip(g1s, g2s):
    betas = np.arange(0.0, 2*np.pi, np.pi/6)
    e1hats = []
    e2hats = []
    for beta in betas:
        e1, e2 = shearme(e*math.cos(2*beta), e*math.sin(2*beta), g1, g2)
        im = drawtrue(e1, e2, HLR, nu, psffwhm)
        e1hat, e2hat = fitme(im, e1, e2, HLR, nu, jmax, psffwhm)
        e1hats.append(e1hat)
        e2hats.append(e2hat)
    g1hats.append(np.mean(e1hats))
    g2hats.append(np.mean(e2hats))
c1 = g1hats[0]
c2 = g2hats[0]
m1 = (g1hats[1]-c1)/g1s[1]-1.0
m2 = (g2hats[1]-c2)/g2s[1]-1.0
print((" {:>9s}"*4).format("m1","m2","c1","c2"))
print((" {:9.5f}"*2 + " {:9.5f}"*2).format(m1, m2, c1, c2))
```
