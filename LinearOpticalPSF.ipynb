```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
import numpy as np
import galsim
```

```python
def img_and_colorbar(fig, grid):
    inner_grid = gs.GridSpecFromSubplotSpec(100, 100, subplot_spec=grid, wspace=0, hspace=0)
    img_ax = fig.add_subplot(inner_grid[:,0:90])
    cbar_ax = fig.add_subplot(inner_grid[:,91:95])
    return img_ax, cbar_ax

def plot(array1, array2, vmin=-4, vmax=0):
    fig = plt.figure(figsize=(15,3))
    grid = gs.GridSpec(1,17)

    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,0:5])
    im = img_ax.imshow(np.log10(array1), vmin=vmin, vmax=vmax)
    plt.colorbar(im, cax=cbar_ax)
    img_ax.set_title("Log10(Series PSF)")

    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,6:11])
    im = img_ax.imshow(np.log10(array2), vmin=vmin, vmax=vmax)
    plt.colorbar(im, cax=cbar_ax)
    img_ax.set_title("Log10(Exact PSF)")

    tmp1 = array1/array1.sum()
    tmp2 = array2/array2.sum()
    img_ax, cbar_ax = img_and_colorbar(fig, grid[0,12:17])
    im = img_ax.imshow((tmp1-tmp2)/tmp2.max(), vmin=-0.2, vmax=0.2)
    plt.colorbar(im, cax=cbar_ax)    
    img_ax.set_title("normalized residual")
```

```python
psf_exact = galsim.OpticalPSF(lam_over_diam=1.0, oversampling=4.0)
psf_series = galsim.LinearOpticalSeries(lam_over_diam=1.0)
img_exact = psf_exact.drawImage(nx=100, ny=100, scale=0.2, method='sb')
img_series = psf_series.drawImage(nx=100, ny=100, scale=0.2, method='sb')
plot(img_series.array, img_exact.array)
```
