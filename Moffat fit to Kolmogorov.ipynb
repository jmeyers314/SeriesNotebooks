```python
%matplotlib inline
import galsim
import matplotlib.pyplot as plt
import numpy as np

betas = np.arange(4.5, 4.91, 0.1)
moffat = [galsim.Moffat(fwhm=1.0, beta=b) for b in betas]
kolmogorov = galsim.Kolmogorov(fwhm=1.0)

colors=['red', 'orange', 'yellow', 'green', 'blue', 'violet']

x = np.logspace(-2, 1, 100)

fig = plt.figure(figsize=(18, 12))
fig.suptitle("Moffat vs. Kolmogorov for fixed FWHM", fontsize=24)

ax = fig.add_subplot(221)
#ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-5, 1)
ax.set_ylabel("$\Sigma(r)$", fontsize=18)
ax.set_xlabel("r", fontsize=18)
for m, b, c in zip(moffat, betas, colors):
    y = [m.xValue(0.0, xi) for xi in x]
    ax.plot(x, y, c=c, label='beta={}'.format(b))
yk = [kolmogorov.xValue(0.0, xi) for xi in x]
ax.plot(x, yk, c='black', label='Kolmogorov')
ax.legend(loc='upper right')

ax = fig.add_subplot(222)
#ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-5, 1)
ax.set_ylabel("$\Sigma(r) - \Sigma_{\mathrm{Kolmogorov}}(r)$", fontsize=18)
ax.set_xlabel("r", fontsize=18)
for m, b, c in zip(moffat, betas, colors):
    y = np.array([m.xValue(0.0, xi) for xi in x]) - np.array(yk)
    ax.plot(x, y, c=c, label='beta={}'.format(b))
    ax.plot(x, -y, '.', c=c)
ax.legend(loc='upper right')

ax = fig.add_subplot(223)
#ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-5, 1)
ax.set_ylabel("$r \Sigma(r)$", fontsize=18)
ax.set_xlabel("r", fontsize=18)

for m, b, c in zip(moffat, betas, colors):
    y = [m.xValue(0.0, xi)*xi for xi in x]
    ax.plot(x, y, c=c, label='beta={}'.format(b))
yk = [kolmogorov.xValue(0.0, xi)*xi for xi in x]
ax.plot(x, yk, c='black', label='Kolmogorov')
ax.legend(loc='upper right')

ax = fig.add_subplot(224)
#ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim(1e-5, 1)
ax.set_ylabel("$r (\Sigma(r) - \Sigma_{\mathrm{Kolmogorov}}(r))$", fontsize=18)
ax.set_xlabel("r", fontsize=18)
for m, b, c in zip(moffat, betas, colors):
    y = np.array([m.xValue(0.0, xi)*xi for xi in x]) - np.array(yk)
    ax.plot(x, y, c=c, label='beta={}'.format(b))
    ax.plot(x, -y, '.', c=c)
ax.legend(loc='upper right')

plt.show()
```
